```python
# Download marketing dataset
df_marketing = pd.read_csv("C:/Users/vishn/Downloads/BostonHousing.csv")
df_marketing
```

```python
# Make sure data types are correct (no objects)
df_marketing
```

```python
# Pop label
target = df_marketing.pop('CAT. MEDV')
```

```python
# Define features
marketing_feature_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'LSTAT', 'MEDV'] 
# Update with relevant feature names
marketing_features = df_marketing[marketing_feature_names]
```

```python
# Convert data to Tensorflow tensor
marketing_features = tf.convert_to_tensor(marketing_features)
marketing_features
```

```python
# Normalize data
normalizer = tf.keras.layers.Normalization(axis=-1)
normalizer.adapt(marketing_features)
```

```python
# Define the model
def get_marketing_model():
  model = tf.keras.Sequential([
    normalizer,
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')  # Binary classification output
  ])

  model.compile(optimizer='adam',
                loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
                metrics=['accuracy'])
  return model
```

```python
# Train model
BATCH_SIZE = 128
model = get_marketing_model()
model.fit(marketing_features, target, epochs=100, batch_size=BATCH_SIZE)
```

```python
# Evaluate model
score = model.evaluate(marketing_features, target, verbose=1)
print(f'Test loss: {score[0]} / Test accuracy: {score[1]}')
```

```python
# Predict labels for new records
df_new_records = pd.read_csv("C:/Users/vishn/Downloads/BostonHousing.csv")
new_records_features = df_new_records[marketing_feature_names]
new_predictions = model(new_records_features, training=False)
```

```python
#display predictions
for i, prob in enumerate(new_predictions):
    label = 1 if prob[0] >= 0.5 else 0
    print(f"Example {i + 1} prediction: Purchase {'Made' if label == 1 else 'Not Made'} (Probability: {prob[0]:.2%})")
```
